```python
import pandas as pd
import nltk

import numpy as np
import matplotlib.pyplot as plt
import gensim
from gensim.models.fasttext import FastText
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
dataset=pd.read_csv('Hindi_Classification_Data.csv')
```

```python
dataset.head(4)
```

```python
hindi=dataset['text']
```

```python
hindi.head(5)
```

```python
hindi[0]
```

```python
len(hindi)
```

```python
hindi[0:5]
```

```python
word_tokenize(hindi[0])
```

```python
y=[nltk.word_tokenize(text) for text in hindi]
```

```python
y[0:10]
```

```python
model = FastText(y, size=300, sg=0, window=10, min_count=5, seed=42, workers=8)
model.save('hindi_fasttext.model')
```
